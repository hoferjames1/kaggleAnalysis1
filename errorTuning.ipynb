```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Your x-values
x_values = [2, 1.5, 0.5, 1, 0.75]

# Initialize an empty list to store means
means_list = []

# Iterate through file names ending with numbers 1 to 5
for i in range(1, 6):
    file_name = f'qualityresults{i}.csv'
    
    # Read the CSV file
    score_data = pd.read_csv(file_name)
    
    # Calculate the mean and append to the list
    mean_value = score_data['Ace Success Ratio'].mean()
    means_list.append(mean_value)

# Plot the means
plt.scatter(x_values, means_list, marker='o', label='Data Points')
plt.xlabel('Distance Coefficient (dc)')
plt.ylabel('Mean Ace Success Ratio (%)')
plt.title('Means for dc 0.5 - 2')

# Fit a polynomial trend line of order 3
coefficients = np.polyfit(x_values, means_list, 3)
poly_trend = np.poly1d(coefficients)

# Generate x values for the trend line
fit_x_values = np.linspace(min(x_values), max(x_values), 100)
# Plot the trend line
plt.plot(fit_x_values, poly_trend(fit_x_values), label='Poly Trend Line (Order 3)')

# Find the x value where y is at the maximum
max_y_index = np.argmax(poly_trend(fit_x_values))
max_x_value = fit_x_values[max_y_index]
max_y_value = poly_trend(max_x_value)

# Mark the maximum point on the plot
plt.scatter(max_x_value, max_y_value, color='red', label='Max Point')

# Show legend
plt.legend()

# Save the plot as an image
plt.savefig('plot_image.png')

# Show the plot
plt.show()

# Print the x value where y is at the maximum
print(f"The x value where y is at the maximum: {max_x_value}")
```
